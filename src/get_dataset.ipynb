```python
sys.path.append("/Users/ruinz/.pyenv/versions/3.9.5/lib/python3.9/site-packages")
import requests
import json
import sys
import pandas as pd
import datetime as dt
import time
```

```python
def get_video_data(month, day, singer):
  filter = json.dumps(
    {
      "type": "or",
      "filters": [
      {
        "type": "range",
        "field": "startTime",
        "from": f"2022-{month}-{day}T00:00:00+09:00",
        "to": f"2022-{month}-{day}T23:59:59+09:00",
        "include_lower": True
      }
      ]
    }
  )
  param = {
      "q":f"{singer}オリジナル曲",
      "targets":"tags",
      "_sort":"-startTime",
      "fields":"title,viewCounter,tags",
      "_limit":100,
      "jsonFilter":filter,
      }
  url = "https://api.search.nicovideo.jp/api/v2/snapshot/video/contents/search"
  data = requests.get(url,params=param).json()
  return data
```

```python
kafu_titles, kafu_view_counts = [], []
miku_titles, miku_view_counts = [], []


def date_range(start, stop, step = dt.timedelta(1)):
    current = start
    while current < stop:
        yield current
        current += step
```

```python

for date in date_range(dt.date(2022,1,1),dt.date(2022,7,1)):
  month = str(date.month).zfill(2)
  day = str(date.day).zfill(2)
  print(f"{month}-{day}")
  miku_got_video_data = get_video_data(month, day, "ミク")
  kafu_got_video_data = get_video_data(month, day, "可不")
  if len(got_video_data) != 0:
    for single_data in miku_got_video_data["data"]:
      miku_titles.append(single_data["title"])
      miku_view_counts.append(single_data["viewCounter"])
  if len(kafu_got_video_data) != 0:
    for single_data in kafu_got_video_data["data"]:
      kafu_titles.append(single_data["title"])
      kafu_view_counts.append(single_data["viewCounter"])
  time.sleep(2)
```

```python
df_miku = pd.DataFrame(list(zip(miku_titles, miku_view_counts)), columns=["Title", "ViewCount"])
df_kafu = pd.DataFrame(list(zip(kafu_titles, kafu_view_counts)), columns=["Title", "ViewCount"])
```
